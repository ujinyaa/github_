```python
import seaborn as sns
import matplotlib.pyplot as plt  
import numpy as np 
import pandas as pd

False_Negative = 2
False_Positive =8
True_Negative = 72
True_Positive = 18

confusion_matrix = np.array([[True_Positive, False_Negative], [False_Positive, True_Negative]])
confusion_df = pd.DataFrame(confusion_matrix, index=["Positive", "Negative"], columns=["True Positive", "False Positive"])

# Plotting the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(confusion_df, annot=True, cmap='coolwarm', fmt='d', cbar=False)
plt.title('Confusion Matrix')
plt.xlabel('Predicted Label')
plt.ylabel('Actual Label')
plt.show()
```

sensitibity

```python
sensitivity = True_Positive / (True_Positive + False_Negative)
```

presicion

```python
precison = True_Positive /(True_Positive+ False_Positive)
```

specificity


```python
specificity = True_Negative / (True_Positive + False_Negative)
```

recall

```python
recall = True_Positive / (False_Negative + True_Positive)
```

F1_SCORE


```python
f1_score = (precison * recall) / (precison+ recall)
```

```python
print(specificity)
print(sensitivity)
print(recall)
print(precison)
print(f1_score)
```
