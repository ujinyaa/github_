```python
top_pop_cities = data.sort_values(by='population_total',ascending=False)
top20_populated_cities=top_pop_cities.head(50)
top20_populated_cities.head(5)
```

```python

data['latitude'] = data['location'].apply(lambda x: x.split(',')[0])
data['longitude'] = data['location'].apply(lambda x: x.split(',')[1])

top20_populated_cities['latitude'] = data.loc[top20_populated_cities.index, 'latitude']
top20_populated_cities['longitude'] = data.loc[top20_populated_cities.index, 'longitude']

top20_populated_cities.head(5)


# data['lattitude'] = data['location'].apply(lambda x: x.split(',')[0])
# data['longitude'] = data['location'].apply(lambda x: x.split(',')[1])

# # 결과 확인
# top20_populated_cities.head(5)
```

```python
from IPython.display import display

top20_female = data.nlargest(50, 'sex_ratio')[['name_of_city','population_total', 'population_female','sex_ratio']]
display(top20_female.head(5))
```

```python
data['city_literatability_ratio'] = data['literates_total'] / data['total_graduates']
top20city_literatability= data.nlargest(50, 'city_literatability_ratio')[['name_of_city', 'literates_total', 'effective_literacy_rate_total', 'total_graduates']]

data['female_literatability_ratio'] = data ['literates_total'] / data['effective_literacy_rate_female']
top20female_literatability = data.nlargest(50,'female_literatability_ratio')[['name_of_city', 'literates_total','literates_female', 'female_graduates']]
display(top20city_literatability.head(5))
display(top20female_literatability.head(5))
```

top20city_literatability = 리터러시 능력 전체 / 졸업  == 공부 비율에 도시 사는 수준 비교  
top20female_literatability = 리터러시 능력 전체 / 리터러시 가능한 여자  == 도시에서의 여성 중요도 파악가능
