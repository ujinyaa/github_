```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = "C:\\Users\\LYJ\\Desktop\\CODE_HOME\\리메셀\\india economic\\1 - India_Economic_Growth_Historical_Data.csv"
data = pd.read_csv(file_path)

# Display the first few rows of the dataset
data.head()
```

```python
# Drop the 'Unnamed: 0' column
data.drop(columns=['Unnamed: 0'], inplace = True)

# Function to clean and convert columns to numeric
def clean_column(column):
    return pd.to_numeric(column.replace('[\$,B%]', '', regex=True))

# Clean and convert the columns
data['GDP'] = clean_column(data['GDP'])
data['Per_Capita'] = clean_column(data['Per_Capita'])
data['Annual_percent_geowth'] = clean_column(data['Annual_percent_geowth'])

# Verify the data types after cleaning
data.info()

# Display the cleaned data
data.head()
```

```python
data.describe()
```

```python
plt.figure(figsize=(12, 6))


sns.lineplot(x='year', y='GDP', data=data, marker='o', label='GDP of India Over Time')
sns.lineplot(x='year', y='Per_Capita', data=data, marker='o', color='orange', label='Per Capita Income')

plt.title("GDP and Per Capita Income of India Over Time")
plt.xlabel('Year')
plt.ylabel('Value')
plt.grid(True)

plt.legend()

plt.show()
```
